### 3) 데이터 추출

#### (1) Species(종)이 setosa만 추출

```python
print("-------------- Species이 setosa인 것만 추출 --------------")

df_setosa=(df["Species"]=="setosa")
display(df[df_setosa])
```

#### (2) setosa에서 꽃받침길이(Sepal length)가 5이하인 데이터만 추출

```python
print("-------- 꽃받침길이(Sepal length)가 5이하인 데이터만 추출 --------")

df_Slen=(df["Sepal length"]<=5)
display(df[df_Slen])
```

#### (3) setosa에서 꽃받침 폭 속성만 추출

```python
print("-------- setosa에서 꽃받침 폭 속성만 추출 --------")

df_width=df.drop(["Sepal length","Petal length"],axis=1) #해당 안되는 속성은 drop
display(df_width)
```

#### (3) setosa에서 꽃받침 길이 속성만 추출

```python
print("-------- setosa에서 꽃받침 길이 속성만 추출 --------")

df_length=df.drop(["Sepal width","Petal width"],axis=1) #해당 안되는 속성은 drop
display(df_length)
```

### 4) 데이터 추가

#### (1) 꽃받침 길이( Sepal length)와 꽃받침 폭( Sepal width)의 비율을 새로운 속성으로 추가

```python
print("----------------------- 새로운 속성으로 추가(Sepal) -----------------------")

df_Sepal=df.copy()
df_Sepal["Sepal ratio"]=df["Sepal length"]/df["Sepal width"]
display(df_Sepal)
```

#### (2) 꽃잎 길이(Petal length) 와 꽃잎 폭(Petal width)의 비율을 새로운 속성으로 추가

```python
print("----------------------- 새로운 속성으로 추가(Petal) -----------------------")

df_Petal=df.copy()
df_Petal["Petal ratio"]=df["Petal length"]/df["Petal width"]
display(df_Petal)
```

### 5) 그룹별 분리, 샘플링, 시각화

#### (1) Species을 기준으로 groupby 처리

```python
grouped=df.groupby("Species")
```

#### (2) Species별 평균, 표준편차 확인

```python
print("---------------------- 평균 확인 ----------------------")
display(grouped.mean())
```

```python
print("-------------------- 표준편차 확인 --------------------")
display(grouped.std()) #standard deviation
```

#### (3) Species(종별) 로 분리

```python
setosa = grouped.get_group(name="setosa")
versicolor = grouped.get_group(name="versicolor")
virginica = grouped.get_group(name="virginica")

print("------------------------ setosa ------------------------")
display(setosa)
print("------------------------ versicolor ------------------------")
display(versicolor)
print("------------------------ virginica ------------------------")
display(virginica)
```

#### (4) 분리된 3개의 종에서 각각 20개씩 랜덤 추출하여 결합

```python
print("-------- 분리된 3개의 종에서 각각 20개씩 랜덤 추출하여 결합 --------")
se_20=setosa.sample(20)
ver_20=versicolor.sample(20)
vir_20= virginica.sample(20)

iris20=se_20.append(ver_20)
iris20=iris20.append(vir_20)

display(iris20)
```

### (5) 3개 종에 대한 boxplot (subplot 이용)

### (6) 3개 종에 대한 histogram

```python
# 아래 histogram 출력시 함수를 호출하여 사용
def set(name,x,y):
    plt.title(name)
    plt.xlabel(x)
    plt.ylabel(y)
```

### (8) (4)의 결합된 데이터의 scatter matrix

```python
import mglearn
```

```python
iris_class = iris20["Species"].map({"setosa": 0,"versicolor": 1,"virginica": 2})

# Kde
pd.plotting.scatter_matrix(iris20,s=60,figsize=(15, 15),marker='o',c=iris_class,
                           diagonal='kde',cmap = mglearn.cm3)

# Histogram
pd.plotting.scatter_matrix(iris20,s=60,figsize=(15, 15),marker='o',c=iris_class,
                           diagonal='hist',cmap = mglearn.cm3)
```
