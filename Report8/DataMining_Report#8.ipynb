#### 1) 데이터 시각화의 중요성에 대하여 Anscombe's quartet로 설명하세요.  
#### 파이썬 코드를 작성하여 설명하면 추가점수 부여함.

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
import statistics
```

```python
#출처: http://seaborn.pydata.org/examples/anscombes_quartet.html
df = sns.load_dataset("anscombe")
display(df)
```

```python
mean_x = df['x'].mean()
sample_x = statistics.variance(df['x'])
mean_y = df['y'].mean()
sample_y = statistics.variance(df['y'])
corr = df['x'].corr(df['y'])

print("Mean of x =",mean_x)
print("Sample variance of x =",sample_x)
print("Mean of y =",mean_y)
print("Sample variance of x =",sample_y)
print("Correlation between x and y =",corr)
```

```python
sns.lmplot(x="x", y="y", col="dataset", hue="dataset", data=df,
           col_wrap=2, ci=None, palette="muted", height=4,
           scatter_kws={"s": 50, "alpha": 1})
```

### Bar plot
범주형에 따른 개수, 퍼센트 등을 이용하여  막대그래프로 시각화  

```python
#단일값 그래프
df.plot.bar(x='x', y='y', title="Bar plot",rot=0)
plt.show()

#다중값 그래프
display(df)
df.plot.bar(rot=0)
plt.show()

#세부속성 y값 순서, x축 값 기울기 조정
df.plot.bar(y="y", rot=45) 
plt.show()

#단일 y값 출력
df.plot.bar(y="y", rot=0) 
plt.show()

#subplot으로 분리된 그래프
axes = df.plot.bar(rot=0, subplots=True,  figsize=(8, 4))
plt.show()
```

### Pie plot

범주형에 따른 개수, 퍼센트 등을 이용하여  원그래프로 시각화  


```python
# x 그래프
df.plot.pie(y='x', title = 'pie plot (x)')  #x만 그리기
# y 그래프
df.plot.pie(y='y', title = 'pie plot (y)',  autopct='%.2f%%')  #y만, % 삽입 그리기
```

### histogram
특정 구간에 속하는 데이터의 빈도수(frequency)를 막대그래프로 표현  
수치 데이터의 분포를 표현

```python
#histogram
df.plot.hist()
df.plot.hist( subplots=True)
df.plot.hist(y='x')
df.plot.hist(y='x', bins=3)

#다중 histogram 
df.hist()
df.hist(bins=3)

plt.show()
```

### density  
  확률밀도함수 Probability Density  Function, pdf)  
  커널밀도추정( kernel density estimation,KDE) 

```python
df.hist(bins=3)

#Generate Kernel Density Estimate plot using Gaussian kernels.
df.plot.density()
df.plot.kde()
plt.show()
```

```python
#plt를 이용한 Histogram, 다양한 그래픽 요소 추가 

mu, sigma = 100, 15
x = np.random.normal(mu, sigma, size=1000) #평균 60, 표준편차 20인 정규분포로부터 1000개 난수생성

n, bins, patches = plt.hist(x, bins=10)
plt.xlabel('x')
plt.ylabel('Probability')
plt.title('Histogram1')
plt.text(120, 200, '$ \mu=100,\ \sigma=15 $') 
plt.show()

print ("n=", n)   #빈도수
print ("bins=", bins) #구간


sx = pd.Series(x) #x를 이용하여 pandas의 Series로 생성

sx.plot.hist()
plt.xlabel('x')
plt.ylabel('Probability')
plt.title('Histogram2')
plt.text(120, 200, '$ \mu=100,\ \sigma=15 $') 
plt.show()

sx.plot.kde()
plt.xlabel('x')
plt.ylabel('Probability')
plt.title('Histogram2')
plt.show()
```

### boxplot
   사 분위수(quartiles)를 통해 수치 데이터 분포를 확인하는 그래프 

```python
df = pd.DataFrame(np.random.randn(10,4), columns=['A','B', 'C', 'D'])

display(df)
df.boxplot()
df.plot.density()
plt.show()
```

### 산점도(Scatter plot)
두 수량형 변수의 값을 x, y 좌표로 하는 점으로 표현  
X값을 입력변수, y값을 출력변수로 관계 확인  
두 변수의 선형적 상관관계 분석에 활용  

```python
# 속성 2개 비교
display(df)

df.plot.scatter(x='x',  y='y',   c='Blue')  #c로 색상지정
df.plot.scatter(x='x',  y='y', colormap='plasma') #type을 칼라맵의 인덱스로 지정
df.plot.scatter(x='x',  y='y',   marker='^', s= 60 ) #마커모양과 크기 지정

plt.show()
```

### Scatter Matrix
여러 변수에 대한 scatter plot의 matrix

```python
pd.plotting.scatter_matrix(df, s=60, diagonal='kde', cmap='spring') #class로 spring색상 설정
pd.plotting.scatter_matrix(df, s=60, diagonal='hist', cmap= 'viridis')
plt.show()
```

### 상관계수 (correlation coefficient) 

두 변수 사이의 선형적인 상관관계를  측정   
두 변수 X 와 Y 간의 선형 상관 관계를 계량화한 수치  
-1~1사이 값을 가지며, 1또는 -1에 가까울수록 상관관계가 강하고, 0에 가까울 수록 약함. 


```python
print("df.corr= \n",df.corr())
```

### 데이터 시각화의 중요성에 대하여

1. Anscombe's quartet의 데이터셋을 가지고 데이터 분석을 진행했다. 데이터 시각화에 있어서 유명한 라이브러리인 matplotlib.pyplot모듈을 이용했다. Simple plot, Bar plot, Pie plot, histogram, density, boxplot, Scatter plot, Scatter Matrix, Correlation coefficient를 통해 그래프를 확인할 수 있었다. 이를 통해 기술통계량은 유사하지만 분포나 그래프는 매우 다르다는 것을 그래프를 통해 한눈에 확인할 수 있었다.

</br>

2. 수치로 표현되는 통계량은 체감상 어떤 수인지 쉽게 알 수 없어 함수 및 수식을 이용해서 확인해야 했지만, 그래프를 통해 시각화된 통계량은 한눈에 통계량을 확인할 수 있었다. 또한 많은 양의 데이터를 한눈에 볼 수 있고 시각적 요소를 활용해 데이터 요약 및 표현을 해준다는 점이 데이터 시각화가 필요한 이유라는 생각이 들었다.

</br>

3. 즉, 데이터 시각화를 통해 데이터가 의미하는 바를 쉽게 찾을 수 있으며 그래프에 표현된 크기, 위치, 색을 통해 데이터를 비교하고 분포를 파악하거나 데이터 안에 숨겨진 유의미한 것을 찾음으로써 데이터 시각화가 데이터 분석에 있어서 효과적인지 체감할 수 있었다.
