### 1. physicalFitness문제

```python
import pandas as pd
import numpy as np
from IPython.display import display
```

### (1)ageGroup(1-9),sex(1,2),height,weight,BMI,bodyFat으로 구성된 파일 (physicalFitness.csv)를 읽고 데이터프레임 만들기

```python
df=pd.read_csv("physicalFitness.csv")

display(df)
```

### (2) (1)의 데이터 셋에 대한 데이터 타입들, 요약통계, 상,하위 5개 데이터 확인, 행 레이블, 열레이블 , 크기 확인

```python
print("-------------------- 데이터 타입 확인 --------------------")
display(df.describe())

print("-------------------- 요약 통계 확인 --------------------")
display(df.describe())

print("-------------------- 상위 5개 데이터 확인 --------------------")
display(df.head(5))

print("-------------------- 하위 5개 데이터 확인 --------------------")
display(df.tail(5))

print("-------------------- 행 레이블 확인 --------------------")
display(df.index)

print("-------------------- 열 레이블 확인 --------------------")
display(df.columns)

print("-------------------- 크기 확인 --------------------")
display(df.shape)
```

### (3) 키가 큰 순으로 정렬

```python
a=df.sort_values(by="height",ascending=False)
display("---------- 키가 큰 순으로 정렬 ----------",a)
```

### (4) BMI 30이상인 비만자만 ageGroup,height,weight,BMI,bodyFat을 추출하여 파일(obese.csv)로 저장

```python
BMI=df['BMI']>=30
BMI_Print=df[BMI]
#display(BMI_Print)

BMI_30=BMI_Print.loc[:,['ageGroup','height','weight','BMI','bodyFat']]
BMI_30.to_csv("obese.csv",mode='w')
```

### (5) (4)에서 추출된 데이터셋에서 나이그룹 속성으로 그룹을 분리

```python
Group=BMI_30.groupby('ageGroup')
for name, group in Group:
    display("----------------- 그룹 %d -----------------" %(name))
    display(group)
```

### (6) (5)그룹에 대한 요약 통계

```python
r1 = Group.agg(['sum', 'mean', 'std'])
display(r1)
```

### (7) 남녀 각각 20명씩 추출하여 하나의 데이터 프레임으로 결합 , 요약통계확인

```python
M=df[df.sex==1].sample(n=20)
F=df[df.sex==2].sample(n=20)
MF=M.append(F)
display("------- 남녀 20명 추출후 데이터 프레임 결합 --------",MF)

print("------------------------ 요약 통계 확인 ------------------------")
MF.describe()
```
